```python
import os
from pathlib import Path

import cv2
import numpy as np
from matplotlib import pyplot as plt

from keras_models import IMAGE_SHAPE
from utils.data_utils import load_image
from utils.traj_utils import brightness_perturbation, noise_perturbation, contrast_perturbation, \
    saturation_perturbation, darkness_perturbation
```

```python
IMG_PATH = "/media/dolphonie/Data/Files/UROP/devens_data/04-21-22_devens_2/1650560180.64_ncp_fine_train/1650560197.732.png"
SAVE_DIR = "/home/dolphonie/Desktop/perturb_images"
Path(SAVE_DIR).mkdir(parents=True, exist_ok=True)
```

```python
image = load_image(IMG_PATH, IMAGE_SHAPE, reverse_channels=True)
image = np.squeeze(image, axis=0)
```

```python
plt.imshow(image)
```

```python
BRIGHT_DELTAS = [0.1, 0.2, 0.3, .4]
for i, delta in enumerate(BRIGHT_DELTAS):
    mod_img = brightness_perturbation(image, delta).numpy()
    np.clip(mod_img, 0, 255)
    plt.subplot(1, 8, i+1)
    plt.imshow(mod_img)
    rev_img = mod_img[..., ::-1]
    cv2.imwrite(os.path.join(SAVE_DIR, f"bright_{delta}.png"), rev_img)
```

```python
DARK_DELTAS = [0.1, 0.2, 0.3, .4]
for i, delta in enumerate(BRIGHT_DELTAS):
    mod_img = darkness_perturbation(image, delta).numpy()
    np.clip(mod_img, 0, 255)
    plt.subplot(1, 8, i+1)
    plt.imshow(mod_img)
    rev_img = mod_img[..., ::-1]
    cv2.imwrite(os.path.join(SAVE_DIR, f"dark_{delta}.png"), rev_img)
```
